```python
# TASK 3: Calculate the distances between a launch site to its proximities
def haversine_distance(lat1, lon1, lat2, lon2):
    """
    Calculate the great-circle distance between two points 
    on the Earth specified in decimal degrees.
    """
    R = 6371.0  # Earth's radius in kilometers

    # Convert decimal degrees to radians
    phi1, phi2 = np.radians(lat1), np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)

    # Haversine formula
    a = np.sin(delta_phi / 2.0)**2 + \
        np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))

    distance = R * c
    return distance
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
launch_site_lat = 28.562302
launch_site_lon = -80.577356
def find_closest_coastline(launch_lat, launch_lon, coastline_df):
    """
    Find the closest coastline point to the launch site.
    Args:
        launch_lat (float): Latitude of the launch site.
        launch_lon (float): Longitude of the launch site.
        coastline_df (DataFrame): DataFrame with 'Lat' and 'Lon' columns for coastline points.
    Returns:
        tuple: Closest coastline coordinates (latitude, longitude).
    """
    min_distance = float('inf')
    closest_point = (None, None)
    
    for _, record in coastline_df.iterrows():
        dist = calculate_distance(launch_lat, launch_lon, record['Lat'], record['Lon'])
        if dist < min_distance:
            min_distance = dist
            closest_point = (record['Lat'], record['Lon'])
    
    return closest_point
# Example DataFrame with coastline points
import pandas as pd

coastline_data = {
    'Lat': [28.56367, 28.56456, 28.56545],
    'Lon': [-80.57163, -80.57054, -80.56945]
}
coastline_df = pd.DataFrame(coastline_data)

# Find closest coastline point
closest_coastline = find_closest_coastline(launch_site_lat, launch_site_lon, coastline_df)
print(f"Closest coastline coordinates: {closest_coastline}")
```

```python

import folium
from folium.plugins import MarkerCluster
from folium.features import DivIcon
import pandas as pd

def find_closest_coastline(launch_lat, launch_lon, coastline_df):
    min_distance = float('inf')
    closest_point = (None, None)
    for _, record in coastline_df.iterrows():
        dist = calculate_distance(launch_lat, launch_lon, record['Lat'], record['Lon'])
        if dist < min_distance:
            min_distance = dist
            closest_point = (record['Lat'], record['Lon'])
    
    return closest_point, min_distance
# Example DataFrame with coastline points
coastline_data = {
    'Lat': [28.56367, 28.56456, 28.56545],
    'Lon': [-80.57163, -80.57054, -80.56945]
}
coastline_df = pd.DataFrame(coastline_data)
# Example launch site coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Find the closest coastline point and distance
closest_coastline, distance_to_coast = find_closest_coastline(launch_site_lat, launch_site_lon, coastline_df)

# Create base map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add a marker for the launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='blue'),
    popup="Launch Site"
).add_to(site_map)

# Add a marker for the closest coastline point
folium.Marker(
    location=closest_coastline,
    icon=folium.Icon(color='green'),
    popup=f"Closest Coastline\nDistance: {distance_to_coast:.2f} KM"
).add_to(site_map)

# Add a distance marker using DivIcon
distance_marker = folium.Marker(
    location=closest_coastline,
    icon=DivIcon(
        icon_size=(150, 30),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12px; color: #d35400;"><b>{distance_to_coast:.2f} KM</b></div>'
    )
)
distance_marker.add_to(site_map)
coordinates = [[launch_site_lat, launch_site_lon], list(closest_coastline)]
lines = folium.PolyLine(
    locations=coordinates,
    weight=2,
    color='red',
    dash_array='5, 5'
).add_to(site_map)
# Display the map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
def find_closest_point(launch_lat, launch_lon, points_df):
    min_distance = float('inf')
    closest_point = (None, None)
    
    for _, record in points_df.iterrows():
        dist = calculate_distance(launch_lat, launch_lon, record['Lat'], record['Lon'])
        if dist < min_distance:
            min_distance = dist
            closest_point = (record['Lat'], record['Lon'])
    
    return closest_point, min_distance

# Sample Data for City, Railway, and Highway
data = {
    "City": {
        'Lat': [28.6101, 28.4000, 28.4500],
        'Lon': [-80.6557, -80.5800, -80.6100]
    },
    "Railway": {
        'Lat': [28.56367, 28.5900],
        'Lon': [-80.57163, -80.6100]
    },
    "Highway": {
        'Lat': [28.5621, 28.5800],
        'Lon': [-80.5732, -80.6200]
    }
}

# Convert to DataFrames
city_df = pd.DataFrame(data["City"])
railway_df = pd.DataFrame(data["Railway"])
highway_df = pd.DataFrame(data["Highway"])

# Launch Site Coordinates
launch_site_lat = 28.562302
launch_site_lon = -80.577356

# Find Closest Points and Distances
closest_city, distance_to_city = find_closest_point(launch_site_lat, launch_site_lon, city_df)
closest_railway, distance_to_railway = find_closest_point(launch_site_lat, launch_site_lon, railway_df)
closest_highway, distance_to_highway = find_closest_point(launch_site_lat, launch_site_lon, highway_df)

# Create base map
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=13)

# Add a marker for the launch site
folium.Marker(
    location=[launch_site_lat, launch_site_lon],
    icon=folium.Icon(color='blue'),
    popup="Launch Site"
).add_to(site_map)

# Function to add markers and lines
def add_marker_and_line(map_object, location, distance, label, color):
    # Add marker with distance label
    folium.Marker(
        location=location,
        icon=DivIcon(
            icon_size=(150, 30),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 12px; color: {color};"><b>{label}: {distance:.2f} KM</b></div>'
        )
    ).add_to(map_object)
    
    # Draw line
    folium.PolyLine(
        locations=[[launch_site_lat, launch_site_lon], location],
        weight=2,
        color=color,
        dash_array='5, 5'
    ).add_to(map_object)

# Add markers and lines for each category
add_marker_and_line(site_map, closest_city, distance_to_city, "City", "green")
add_marker_and_line(site_map, closest_railway, distance_to_railway, "Railway", "red")
add_marker_and_line(site_map, closest_highway, distance_to_highway, "Highway", "orange")

# Display the map
site_map
```
